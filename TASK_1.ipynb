```python
# Step 1: Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Step 2: Load dataset
df = pd.read_csv('World Population Growth.csv')
```

```python
# Step 3: Clean and convert relevant columns to numeric
df['Population'] = df['Population'].str.replace(',', '').astype(float)
df['Number'] = df['Number'].str.replace(',', '').astype(float)
df['Yearly Growth %'] = df['Yearly Growth %'].str.replace('%', '').astype(float)
```

```python
# Step 4: Bar Graph - Population by Year (Top 20 Years)
plt.figure(figsize=(12, 6))
top_years = df.sort_values(by='Population', ascending=False).head(20)
sns.barplot(data=top_years, x='Year', y='Population', palette='viridis')
plt.xticks(rotation=45)
plt.title('Top 20 Years by Population')
plt.xlabel('Year')
plt.ylabel('Population')
plt.tight_layout()
plt.show()
```
